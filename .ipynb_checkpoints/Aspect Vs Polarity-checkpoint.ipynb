```python
import collections
import glob
import json
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.max_colwidth', None, 'display.max_rows', None, )

dataset_dir = "./datasets/"

#SUBSETS = "train dev test".split()
SUBSETS = ["train"]


datasets = collections.defaultdict(list)

for subset in SUBSETS:
    for filename in glob.glob(dataset_dir + subset + "/*"):
        with open(filename, 'r') as f:
            datasets[subset].append(json.load(f))
            
all_pairs = sum(datasets.values(), [])

def total_and_average_len(list_of_lists):
    big_list = sum(list_of_lists, [])
    return len(big_list), len(big_list)/len(list_of_lists)

def count_dataset(pairs, subset):
    # TODO: Add double-annotated and adjudicated
    review_total, review_average = total_and_average_len([pair["review_sentences"] for pair in pairs])
    rebuttal_total, rebuttal_average = total_and_average_len([pair["rebuttal_sentences"] for pair in pairs])
    return {
        "subset":subset,
        "pairs": len(pairs),
        "forums": len(set(pair["metadata"]["forum_id"] for pair in pairs)),
        "adjudicated": len([pair for pair in pairs if pair["metadata"]["annotator"] == "anno0"]),
        "review_sentences": review_total,
        "rebuttal_sentences": rebuttal_total,
        "review_avg_sentences": review_average,
        "rebuttal_avg_sentences": rebuttal_average,
        
    }
# Distribution of examples over sets
df_dicts = [count_dataset(pairs, subset) for subset, pairs in datasets.items()]
df = pd.DataFrame.from_dict(df_dicts)
df.head()
```

```python
df = pd.DataFrame({'review_id':[],'rating':[], 'sentence_index':[], 'text':[], 'coarse':[], 'fine':[], 'asp':[], 'pol':[]},
                  columns = ['review_id', 'sentence_index', 'text', 'coarse', 'fine', 'asp', 'pol'])
for pair in all_pairs:
    rs = pd.json_normalize(pair, record_path=['review_sentences'])
    rs['rating'] = pair['metadata']['rating']
    df = df.append(rs)

df.head()
```

### Aspect Vs. Polarity

```python
## since values with aspect=none does not have polarity: pol, value_count
## none            2182
## pol_negative       2
# pivot.drop(inplace=True, index=pivot.loc[pivot.asp=='none'].index)
# pivot = pivot.rename(columns={'index':'count'})
# pivot.head()
```

```python
sns.set(rc={'figure.figsize':(5,4)})

# fig, ax = plt.subplots(4, 3)
for r in df.rating.unique():
    r = int(r)
    pivot = df.loc[df.rating == r].groupby(['asp', 'pol'], as_index=False).agg({'text':'count'})
    pivot.drop(inplace=True, index=pivot.loc[pivot.asp=='none'].index) # remove aspect None for more clarity on other aspects
    pivot = pivot.rename(columns={'text':'count'})
    pivot['count'] = pivot['count']/pivot['count'].max()
    
    plt.subplot(4,3, r-1)
    __ = sns.barplot(x='asp', y='count', data=pivot, hue='pol')
    _ = plt.xticks(rotation=90)
    ___ = plt.title(f'Aspect frequency by polarity: Rating {r}')
```

# Aspect comparison across ratings

```python
pos_df = df.loc[(df.pol == 'pol_positive') & (df.asp.isin(['asp_motivation-impact', 'asp_originality', 'asp_clarity', 'asp_substance']))]
pos_df_ = pos_df.groupby(['rating', 'asp'], as_index=False).agg({'text': 'count'}).rename(columns={'text':'count'}).astype({'count':'float64'})
pos_df_max_counts = df.loc[(df.pol == 'pol_positive')].groupby(['rating', 'asp'], as_index=False).agg({'text': 'count'}).rename(columns={'text': 'upper_lmt'})
pos_df_max_counts = pos_df_max_counts.groupby(['rating']).agg({'upper_lmt': 'max'})
for idx, row in pos_df_.iterrows():
    pos_df_['count'][idx] = np.round(pos_df_['count'][idx]/pos_df_max_counts.upper_lmt[pos_df_.rating[idx]],2)

sns.set(rc={'figure.figsize':(10,8)})

sns.lineplot(x='rating', y='count', hue='asp', data=pos_df_)
plt.legend(loc='upper right', title='asp') 
```

```python
neg_df = df.loc[(df.pol == 'pol_negative') & (df.asp.isin(['asp_soundness-correctness', 'asp_clarity', 'asp_substance']))]
pos_df = neg_df

pos_df_ = pos_df.groupby(['rating', 'asp'], as_index=False).agg({'text': 'count'}).rename(columns={'text':'count'}).astype({'count':'float64'})
pos_df_max_counts = df.loc[(df.pol == 'pol_negative')].groupby(['rating', 'asp'], as_index=False).agg({'text': 'count'}).rename(columns={'text': 'upper_lmt'})
pos_df_max_counts = pos_df_max_counts.groupby(['rating']).agg({'upper_lmt': 'max'})
for idx, row in pos_df_.iterrows():
    pos_df_['count'][idx] = np.round(pos_df_['count'][idx]/pos_df_max_counts.upper_lmt[pos_df_.rating[idx]],2)

sns.set(rc={'figure.figsize':(10,8)})

sns.lineplot(x='rating', y='count', hue='asp', data=pos_df_)
plt.legend(loc='upper right', title='asp') 
plt.title('Rating vs Aspect frequencies')
```
